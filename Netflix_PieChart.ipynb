```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Create a pie chart which shows the 5 top most popular genres of the movies in the dataset

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df
```

```python
movies = pd.DataFrame(df[df['type'] == 'Movie'])
movies.head(5)
```

```python
def to_list(x):
    x = x.split(",")

    new_x = []
    for i in x:
        i = i.strip()
        new_x.append(i)
    return new_x
```

```python
movies['listed_in'] = movies['listed_in'].apply(to_list)
```

```python
movies
```

```python
genres = movies['listed_in'].explode().unique()
genres
```

```python
genres_list = []

for g in list(movies['listed_in']):
    for i in g:
        genres_list.append(i)
```

```python
#genres_list
```

```python
genre_counts = []

for i in genres_list:
    genre_counts.append(genres_list.count(i))
```

```python
#genre_counts
```

```python
genres_df = pd.DataFrame(list(genres_list), columns=['Genre'])
count_df = pd.DataFrame(list(genre_counts), columns=['Count'])

genre_df = pd.concat([genres_df, count_df], axis=1)
```

```python
genre_df
```

```python
genre_df['Genre'].unique()
```

```python
# using mean will give the actual numbers since every instance is in the dataframe and we are using groupby
genre_df.groupby('Genre')['Count'].mean().sort_values(ascending=False)
```

```python
# top 5 are International Movies, Dramas, Comedies, Documentaries, Action & Adventure
```

```python
# getting top 5 genres as percentage of 6131
top_5 = ['International Movies', 'Dramas', 'Comedies', 'Documentaries', 'Action & Adventure']
```

```python
counts = [2752, 2427, 1674, 869, 859]
percent = []

for c in counts:
    percent.append((c/6131)*100)
```

```python
# plotting pie chart
```

```python
plt.pie(x=percent, labels=top_5)
```
